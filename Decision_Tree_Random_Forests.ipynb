## Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = cleaned_data.drop('point_count', axis=1)  # features
y = cleaned_data['point_count']  # Itarget variable
```

```python
# Split the dataset into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=999)

# Initialize the Decision Tree Regressor
dt_regressor = DecisionTreeRegressor(random_state=999)

# Fit the regressor to the training data
dt_regressor.fit(X_train, y_train)

# Predict on the test data
y_pred = dt_regressor.predict(X_test)
```

```python
# Evaluate the decision tree regressor
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the results
print('Decision Tree Regressor - Mean Squared Error:', mse)
print('Decision Tree Regressor - R2 Score:', r2)
```

```python
print(f"Max Depth: {dt_regressor.max_depth}")
print(f"Min Samples Split: {dt_regressor.min_samples_split}")
print(f"Min Samples Leaf: {dt_regressor.min_samples_leaf}")
print(f"Cost Complexity Pruning Alpha: {dt_regressor.ccp_alpha}")
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from scipy.stats import randint



# Define models to compare
models = {
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor()
}

# Perform randomized search
param_dist = {
    'n_estimators': randint(10, 200),
    'max_depth': randint(2, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

random_search_rf = RandomizedSearchCV(RandomForestRegressor(), param_distributions=param_dist, n_iter=100, scoring='r2', cv=5, random_state=42)
random_search_gb = RandomizedSearchCV(GradientBoostingRegressor(), param_distributions=param_dist, n_iter=100, scoring='r2', cv=5, random_state=42)

models['Random Search RF'] = random_search_rf
models['Random Search GB'] = random_search_gb

# Perform grid search
param_grid = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [2, 5, 10, 20],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 10]
}

grid_search_rf = GridSearchCV(RandomForestRegressor(), param_grid=param_grid, scoring='r2', cv=5)
grid_search_gb = GridSearchCV(GradientBoostingRegressor(), param_grid=param_grid, scoring='r2', cv=5)

models['Grid Search RF'] = grid_search_rf
models['Grid Search GB'] = grid_search_gb

results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)

    results[name] = {
        'train_mse': mean_squared_error(y_train, y_train_pred),
        'test_mse': mean_squared_error(y_test, y_test_pred),
        'train_r2': r2_score(y_train, y_train_pred),
        'test_r2': r2_score(y_test, y_test_pred),
        'train_mae': mean_absolute_error(y_train, y_train_pred),
        'test_mae': mean_absolute_error(y_test, y_test_pred)
    }

# Display results
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from scipy.stats import randint
from joblib import parallel_backend

# Define models to compare
models = {
    'Decision Tree': DecisionTreeRegressor(random_state=7033),
    'Random Forest': RandomForestRegressor(random_state=7033),
    'Gradient Boosting': GradientBoostingRegressor(random_state=7033)
}

# Perform randomized search for Random Forest
param_dist_rf = {
    'n_estimators': randint(10, 200),
    'max_depth': randint(2, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

random_search_rf = RandomizedSearchCV(RandomForestRegressor(), param_distributions=param_dist_rf, n_iter=100, scoring='r2', cv=5, random_state=42)
models['Random Search RF'] = random_search_rf

# Perform randomized search for Gradient Boosting
param_dist_gb = {
    'n_estimators': randint(10, 200),
    'max_depth': randint(2, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

random_search_gb = RandomizedSearchCV(GradientBoostingRegressor(), param_distributions=param_dist_gb, n_iter=100, scoring='r2', cv=5, random_state=42)
models['Random Search GB'] = random_search_gb

# Perform randomized search for Decision Tree
param_dist_dt = {
    'max_depth': randint(2, 20),
    'min_samples_split': randint(2, 20),
    'min_samples_leaf': randint(1, 20)
}

random_search_dt = RandomizedSearchCV(DecisionTreeRegressor(), param_distributions=param_dist_dt, n_iter=100, scoring='r2', cv=5, random_state=42)
models['Random Search DT'] = random_search_dt

# Perform grid search for Random Forest
param_grid_rf = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [2, 5, 10, 20],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 10]
}

grid_search_rf = GridSearchCV(RandomForestRegressor(), param_grid=param_grid_rf, scoring='r2', cv=5)
models['Grid Search RF'] = grid_search_rf

# Perform grid search for Gradient Boosting
param_grid_gb = {
    'n_estimators': [10, 50, 100, 200],
    'max_depth': [2, 5, 10, 20],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 10]
}

grid_search_gb = GridSearchCV(GradientBoostingRegressor(), param_grid=param_grid_gb, scoring='r2', cv=5)
models['Grid Search GB'] = grid_search_gb

# Perform grid search for Decision Tree
param_grid_dt = {
    'max_depth': [2, 5, 10, 20],
    'min_samples_split': [2, 5, 10, 20],
    'min_samples_leaf': [1, 2, 4, 10]
}

grid_search_dt = GridSearchCV(DecisionTreeRegressor(), param_grid=param_grid_dt, scoring='r2', cv=5)
models['Grid Search DT'] = grid_search_dt

results = {}

for name, model in models.items():
    with parallel_backend('threading'):  # Change to 'loky' or 'multiprocessing' for more options
        model.fit(X_train, y_train)
        y_train_pred = model.predict(X_train)
        y_test_pred = model.predict(X_test)

    results[name] = {
        'train_mse': mean_squared_error(y_train, y_train_pred),
        'test_mse': mean_squared_error(y_test, y_test_pred),
        'train_r2': r2_score(y_train, y_train_pred),
        'test_r2': r2_score(y_test, y_test_pred),
        'train_mae': mean_absolute_error(y_train, y_train_pred),
        'test_mae': mean_absolute_error(y_test, y_test_pred)
    }

# Display results
results_df = pd.DataFrame(results).T
print(results_df)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

model = Sequential()

# Add an Input layer explicitly
model.add(Input(shape=(X_train_scaled.shape[1],)))

# Add the first hidden layer
model.add(Dense(64, activation='relu'))

# Add a second hidden layer
model.add(Dense(32, activation='relu'))

# Add an output layer
model.add(Dense(1)) 
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Fit the ANN to the training set
history = model.fit(X_train_scaled, y_train, batch_size=32, epochs=100, validation_split=0.2)
```

```python
# Predicting the Test set results
y_pred_ann = model.predict(X_test_scaled)
```
