```python
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN

lists = ['SECTION','days','income_log','population_log', 'ln_bldgarea', 'ln_comarea','ln_resarea', 'ln_officearea', 'ln_retailarea',]
temp = data[lists].dropna()

X = temp.iloc[:, 1:]

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=999)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()



n_clusters = 3  
kmeans = KMeans(n_clusters=n_clusters, random_state=999)
temp['cluster_label'] = kmeans.fit_predict(X)

sns.scatterplot(x='population_log', y='income_log', hue='cluster_label', data=temp, palette='viridis')
plt.title('KMeans Clustering')
plt.xlabel('population_log')
plt.ylabel('income_log')
plt.legend(title='Cluster')
plt.show()

cluster_centers = pd.DataFrame(kmeans.cluster_centers_, columns=X.columns)
print("Cluster Centers:")
print(cluster_centers)
```

```python
# by types
lists = ['ownertype_nan', 'ownertype_C', 'ownertype_X', 'ownertype_O', 'ownertype_P', 'ownertype_M', 'days','income_log','population_log']

temp = data[lists]
temp = temp.fillna(0)
temp
```

```python
for l in lists:
    temp[l] = pd.cut(temp[l], bins=5, labels=False)
temp.head()
```

```python
from pgmpy.estimators import HillClimbSearch
from pgmpy.estimators import BicScore

hc = HillClimbSearch(temp)
best_model = hc.estimate(scoring_method='BicScore')
print(best_model.edges())
```

```python
from pgmpy.models import BayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator

model = BayesianNetwork(best_model.edges())
model.fit(temp, estimator=MaximumLikelihoodEstimator)

def LL(x, model, verbose=False):
    loglike = 0
    for cpd in model.get_cpds():
        temp_cpd = cpd.copy()
        thevariable = temp_cpd.variable
        theparents = model.predecessors(thevariable)
        theparents_list = []
        for parent in theparents:
            temp_cpd.reduce([(parent, x[parent])])
            theparents_list.append(parent)
        # Retrieve the state of thevariable from the input data x
        thestate = x[thevariable]
        # Find the index corresponding to the state of the variable in the CPD
        theindex = temp_cpd.state_names[thevariable].index(thestate)
        # Get the probability corresponding to the state
        theprob = temp_cpd.get_values()[theindex]
        if verbose:
            print(thevariable, theparents_list, theprob)
        loglike += np.log(theprob)
    return loglike
```

```python
exmp = temp.apply(lambda x: LL(x, model), axis=1)
```

```python
exmp.sort_values().head(5)
```

```python
exmp = pd.DataFrame(exmp)
head = data.copy()
head = head.merge(exmp, left_index=True, right_index=True)
head.rename(columns={0: 'log-likelihood'}, inplace=True)
head.sort_values('log-likelihood').head(5)
```
