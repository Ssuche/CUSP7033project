```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np


data = pd.read_csv('nyc_pluto_24v1_csv/pluto_24v1_1.csv')
data.head()
```

```python
data['ct2010']
```

```python
count = data.groupby('ct2010').size().reset_index(name='building count')
```

```python
count['ct2010'] = count['ct2010'].astype('float')
```

```python
count
```

```python
data2 = pd.read_csv('data/nyc_census.csv')
data2
```

```python
data2 = data2[['census', 'county', 'state', 'income', 'population', 'tract', 'GEOID20','INTPTLAT20', 'INTPTLON20']]
```

```python
for i in sorted(data2['tract']):
    print(i)
```

```python
data2['census'] = data2['census'].astype('float')
```

```python
for i in sorted(data2['census']):
    print(i)
```

```python
temp = pd.merge(count, data2, left_on = 'ct2010', right_on = 'census', how='inner')
temp
```
